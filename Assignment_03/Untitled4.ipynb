```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv("/content/admission (1).csv")
```

```python
ds.head()
```

```python
#bar plot
plt.bar(ds['University Rating'],ds['TOEFL Score'],color="orange",width=0.6)
plt.title('University Rating vs TOEL Score.')
plt.xlabel('TOEFL Score.')
plt.ylabel('University Rating')
plt.show()
```

```python
#Scatter plot
plt.scatter (ds["TOEFL Score"],ds["CGPA"])
plt.xlabel("TOEFL Score")
plt.ylabel("CGPA")
plt.title("TOEFL Score vs CGPA")
plt.show()
```

```python
sns.boxplot(x="GRE Score",data=ds,width=0.8,showmeans=True)
plt.show()
```

```python
ds.hist(bins=10,figsize=(15,12),layout=(3,3))
plt.suptitle('Histograms for Admission Data',x=0.5,y=0.95,ha='center',fontsize='x-large')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x="TOEFL Score", data = ds, hue = "Research", palette = "rocket")
```

```python
plt.figure(figsize =(15,8))
sns.countplot(x="TOEFL Score", data = ds, color = "grey")
```

```python
research_counts = ds['Research'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(research_counts, labels=['Research (1)', 'No Research (0)'], autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'skyblue'])
```
